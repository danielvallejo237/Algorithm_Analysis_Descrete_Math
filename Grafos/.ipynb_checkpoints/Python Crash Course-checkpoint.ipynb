```python
# OR
print(True or True,True or False, False or True, False or False)
```

```python
#AND 
print(True and True,True and False, False and True, False and False)
```

```python
#IGUALDAD 
print(True is True,True is False, False is True, False is False)
```

## Arreglos

Los arreglos son un tipo de dato capaz de representar elementos en $\mathbb{R}^{n}$. Generalmente usamos la librería $\texttt{numpy}$ para manejar arreglos

```python
#Ejemplo interactivo
import numpy as np
from PIL import Image,ImageOps
import matplotlib.pyplot as plt
imagen=Image.open("./michi_picture.jpg")
imagen=ImageOps.grayscale(imagen)
imagen=imagen.resize((1020,720))
imagen=np.asarray(imagen)
print(imagen.shape)
plt.figure(figsize=(8,8))
plt.imshow(imagen,cmap='gray')
```

```python
def get_matrix(mat,i,j,filtersize):
    return mat[i-filtersize//2:i+filtersize//2+1,j-filtersize//2:j+filtersize//2+1]
def return_feature_map(patch,filtro):
    return sum(sum(patch*filtro,))
def conv_operator(imagen,filtro):
    pila_vec=[]
    for i in range(filtro.shape[0]//2,imagen.shape[0]-filtro.shape[0]//2):
        row_vec=[]
        for j in range(filtro.shape[1]//2,imagen.shape[1]-filtro.shape[1]//2):
            mat=get_matrix(imagen,i,j,filtro.shape[0])
            row_vec.append(return_feature_map(mat,filtro))
        pila_vec.append(row_vec)
    return pila_vec
filtro=np.asarray([[1,0,-1],[2,0,-2],[1,0,-1]])
pv=np.asarray(conv_operator(imagen,filtro))
plt.figure(figsize=(8,8))
plt.imshow(pv,cmap='gray')
plt.title('Detección de esquinas')
```

## Ciclos for y while

Como se describió en las secciones anteriores los ciclos for solo se puede hacer sobre elementos iterables como diccionarios, listas, cadenas, etc... La estructura de un ciclo for es de la forma 
$$
\text{ for } element \text{ in } iterableobject \text{:} \\
\hspace{1cm}\text{Realizar algo}
$$


```python
#Ejemplos de ciclo for con los objetos que tenemos
lista=list(('a','b','c','d','e'))
for el in lista:
    print(el)
```

```python
car={
    'marca':'Ford',
    'modelo':'F-150',
    'año':'2020'
}
```

```python
for element in car:
    print(car[element])
```

```python
for (key,value) in car.items():
    print(key,value)
```

```python
#Tambien tenemos dobles iteradores que se iteran de forma paralela
#Es muy util cuando las listas tienen la misma cantidad de elementos y tenemos una función que depende de los
#dos elementos de la lista
for x,y in zip(lista,lista):
    print(x,y)
```

Un ciclo while funciona con una condición de paro y se sigue ejecutando hasta que dicha condición se satisfaga

```python
while True:
    print("Hello world from Python!")
    break
```

## Funciones

Las funciones en Python comienzan con la palabra reservada def, no se especifica el tipo de dato que se va a regrasar, después de declarar los parámetros necesarios se pone el símbolo : y después se continúa con el contenido de la función

```python
def fibbonacci(n):
    if n==1:
        return 1
    elif n==0:
        return 0
    else:
        return fibbonacci(n-1)+fibbonacci(n-2)
```

```python
fibbonacci(10)
```
